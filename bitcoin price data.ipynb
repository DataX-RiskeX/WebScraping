```python
import pandas as pd
from pandas import DataFrame
import numpy as np
from datetime import datetime,tzinfo
from pytz import timezone
import time
import pytz
```

```python
df_bitstamp = pd.read_csv('bitstampUSD_1-min_data_2012-01-01_to_2018-01-08.csv')
df_bitstamp.head()
```

```python
df_bitstamp.Timestamp = pd.to_datetime(df_bitstamp.Timestamp, unit='s')
df_bitstamp.Timestamp = df_bitstamp.Timestamp.dt.tz_localize('UTC').dt.tz_convert('US/Central')
df_bitstamp.head()
```

```python
price=df_bitstamp['Weighted_Price']
```

```python
min_periods = 10 
# Calculate the volatility
vol = price.rolling(min_periods).std() * np.sqrt(min_periods) 
df_bitstamp['Volatility']=vol
df_bitstamp.head()
```
